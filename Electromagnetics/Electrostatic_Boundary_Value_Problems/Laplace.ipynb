```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

nx, ny, nz = 30, 30, 30
V = np.zeros((nx, ny, nz))

V[:, :, 0] = 100 

for _ in range(5000):
    V[1:-1, 1:-1, 1:-1] = (V[1:-1, 1:-1, :-2] + V[1:-1, 1:-1, 2:] +
                            V[1:-1, :-2, 1:-1] + V[1:-1, 2:, 1:-1] +
                            V[:-2, 1:-1, 1:-1] + V[2:, 1:-1, 1:-1]) / 6

mid = V[:, :, nz//2]

x = np.arange(0, nx)
y = np.arange(0, ny)
X, Y = np.meshgrid(x, y)

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, mid, cmap='plasma')
ax.set_title("Laplaces Equation (3D slice)")
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('Potential (V)')
plt.show()
```
