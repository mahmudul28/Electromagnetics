```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def vector_magnitude(v):
    return np.sqrt(np.sum(v**2))

def unit_vector(v):
    return v / vector_magnitude(v)

def plot_vectors(vectors, colors, labels, title):
    fig = plt.figure(figsize=(7,7))
    ax = fig.add_subplot(111, projection='3d')
    for v, c, label in zip(vectors, colors, labels):
        ax.quiver(0, 0, 0, v[0], v[1], v[2], color=c, arrow_length_ratio=0.1, linewidth=2)
        ax.text(v[0]*1.1, v[1]*1.1, v[2]*1.1, label, color=c, fontsize=10)
    lim = max([np.max(np.abs(v)) for v in vectors]) * 1.3
    ax.set_xlim([-lim, lim])
    ax.set_ylim([-lim, lim])
    ax.set_zlim([-lim, lim])
    ax.set_title(title, fontsize=14, fontweight='bold')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    plt.show()

A = np.array([3, 2, 1], dtype=float)
B = np.array([1, 4, 2], dtype=float)

add = A + B
plot_vectors([A, B, add], ['r', 'b', 'g'], ['A', 'B', 'A+B'], 'Vector Addition')

sub = A - B
plot_vectors([A, B, sub], ['r', 'b', 'g'], ['A', 'B', 'A-B'], 'Vector Subtraction')

dot = np.dot(A, B)
print("A  B =", dot)

cross = np.cross(A, B)
plot_vectors([A, B, cross], ['r', 'b', 'm'], ['A', 'B', 'AB'], 'Vector Cross Product')

div = A / B
plot_vectors([A, B, div], ['r', 'b', 'c'], ['A', 'B', 'A/B'], 'Element-wise Division')

unitA = unit_vector(A)
unitB = unit_vector(B)
plot_vectors([A, unitA], ['r', 'g'], ['A', 'Unit A'], 'Unit Vector of A')
plot_vectors([B, unitB], ['b', 'm'], ['B', 'Unit B'], 'Unit Vector of B')
```
