```python
import numpy as np
import matplotlib.pyplot as plt


t = np.linspace(0, 10, 500)


B0 = 1.0
A0 = 1.0
omega_B = 2 * np.pi * 0.2
omega_A = 2 * np.pi * 0.1


B_t = B0 * np.sin(omega_B * t)
A_const = A0
Phi1 = B_t * A_const
EMF1 = -np.gradient(Phi1, t)


B_const = B0
A_t = A0 * (1 + 0.5 * np.sin(omega_A * t))
Phi2 = B_const * A_t
EMF2 = -np.gradient(Phi2, t)


Phi3 = B_t * A_t
EMF3 = -np.gradient(Phi3, t)


plt.figure(figsize=(10, 6))
plt.plot(t, EMF1, label=r"Case 1: Stationary loop, time-varying B", linewidth=2)
plt.plot(t, EMF2, label=r"Case 2: Time-varying loop area, static B", linewidth=2)
plt.plot(t, EMF3, label=r"Case 3: Time-varying loop area, time-varying B", linewidth=2)

plt.title(r"Induced EMF ($\mathcal{E}$) for Different Cases of Faraday’s Law", fontsize=14)
plt.xlabel("Time (s)", fontsize=12)
plt.ylabel(r"EMF $\mathcal{E}$ (V)", fontsize=12)
plt.legend(fontsize=10)
plt.grid(True)
plt.show()
```
