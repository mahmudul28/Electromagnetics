```python


import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


plt.style.use('dark_background')
fig = plt.figure(figsize=(14, 10))
fig.suptitle("Maxwell’s Equations — The Birth of Electromagnetism", color='white', fontsize=18, fontweight='bold')


def arrow3D(ax, X, Y, Z, U, V, W, color):
    ax.quiver(X, Y, Z, U, V, W, length=0.15, normalize=True, color=color, alpha=0.9)

# Grid
x, y, z = np.meshgrid(np.linspace(-1, 1, 8),
                      np.linspace(-1, 1, 8),
                      np.linspace(-1, 1, 8))


#  Gauss’s Law for Electric Field — ∇·E = ρ/ε₀

ax1 = fig.add_subplot(231, projection='3d')
r = np.sqrt(x**2 + y**2 + z**2) + 1e-3
Ex, Ey, Ez = x/r**3, y/r**3, z/r**3
arrow3D(ax1, x, y, z, Ex, Ey, Ez, 'red')
ax1.set_title("∇·E = ρ/ε₀", color='red', fontsize=12)
ax1.text2D(0.1, 0.95, "Gauss’s Law (Electric Field)", transform=ax1.transAxes, color='white', fontsize=10)
ax1.set_axis_off()


#  Gauss’s Law for Magnetism — ∇·B = 0

ax2 = fig.add_subplot(232, projection='3d')
Bx = -y
By = x
Bz = 0*y
arrow3D(ax2, x, y, z, Bx, By, Bz, 'cyan')
ax2.set_title("∇·B = 0", color='cyan', fontsize=12)
ax2.text2D(0.1, 0.95, "Gauss’s Law (Magnetism)", transform=ax2.transAxes, color='white', fontsize=10)
ax2.set_axis_off()


#  Faraday’s Law — ∇×E = -∂B/∂t

ax3 = fig.add_subplot(233, projection='3d')
Ex = np.sin(np.pi * y)
Ey = np.cos(np.pi * x)
Ez = np.zeros_like(x)
arrow3D(ax3, x, y, z, Ex, Ey, Ez, 'magenta')
ax3.set_title("∇×E = -∂B/∂t", color='magenta', fontsize=12)
ax3.text2D(0.1, 0.95, "Faraday’s Law", transform=ax3.transAxes, color='white', fontsize=10)
ax3.set_axis_off()


#  Ampère-Maxwell Law — ∇×B = μ₀(J + ε₀∂E/∂t)

ax4 = fig.add_subplot(234, projection='3d')
Bx = np.cos(np.pi * y)
By = -np.sin(np.pi * x)
Bz = np.zeros_like(x)
arrow3D(ax4, x, y, z, Bx, By, Bz, 'blue')
ax4.set_title("∇×B = μ₀(J + ε₀∂E/∂t)", color='blue', fontsize=10)
ax4.text2D(0.1, 0.95, "Ampère–Maxwell Law", transform=ax4.transAxes, color='white', fontsize=10)
ax4.set_axis_off()


# Faraday + Ampère — Coupled curls
ax5 = fig.add_subplot(235, projection='3d')
Ex = np.sin(np.pi * y)
Ey = np.cos(np.pi * x)
Bx = np.cos(np.pi * y)
By = -np.sin(np.pi * x)
arrow3D(ax5, x, y, z, Ex, Ey, 0, 'magenta')
arrow3D(ax5, x, y, z, Bx, By, 0, 'blue')
ax5.set_title("Coupled E & B Fields", color='violet', fontsize=11)
ax5.text2D(0.1, 0.95, "Faraday + Ampère", transform=ax5.transAxes, color='white', fontsize=10)
ax5.set_axis_off()


#  All Combined — Full Electromagnetic Wave
ax6 = fig.add_subplot(236, projection='3d')
k = np.pi
Ex = np.sin(k * x)
Ey = np.zeros_like(x)
Ez = np.zeros_like(x)
Bx = np.zeros_like(x)
By = np.sin(k * x + np.pi/2)
Bz = np.zeros_like(x)
arrow3D(ax6, x, y, z, Ex, Ey, Ez, 'red')
arrow3D(ax6, x, y, z, Bx, By, Bz, 'cyan')
ax6.set_title("All Combined → EM Wave", color='orange', fontsize=11)
ax6.text2D(0.1, 0.95, "Unified Electromagnetic Field", transform=ax6.transAxes, color='white', fontsize=10)
ax6.set_axis_off()


plt.tight_layout()
plt.savefig("maxwell_equations_visual.png", dpi=400, bbox_inches='tight', facecolor='black')
plt.show()
```
