```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

h = 0.1
V0 = 100.0
a = 1.0
b = a
c = 4 * V0 / np.pi
IMAX = int(a / h)
JMAX = int(b / h)
NMAX = 13

V = np.zeros((IMAX, JMAX))

for i in range(IMAX):
    x = h * (i + 1)
    for j in range(JMAX):
        y = h * (j + 1)
        sum_val = 0.0
        for n in range(1, NMAX+1, 2):  # odd n only
            a1 = np.sin(n * np.pi * x / b)
            a2 = np.sinh(n * np.pi * y / b)
            a3 = n * np.sinh(n * np.pi * a / b)
            sum_val += c * a1 * a2 / a3
        V[i, j] = sum_val

X = h * (np.arange(IMAX) + 1)
Y = h * (np.arange(JMAX) + 1)
X, Y = np.meshgrid(X, Y, indexing='ij')

fig = plt.figure(figsize=(9,7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, V, cmap='inferno')

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("Potential V")
ax.set_title("Potential Distribution by Fourier Series")

plt.show()
```
