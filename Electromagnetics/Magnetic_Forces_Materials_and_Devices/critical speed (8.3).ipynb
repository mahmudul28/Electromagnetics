```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

E_mag = 20.0
B_mag = 5.0
v_mag = 4.0

u_c = E_mag / B_mag

speeds = [0.8*u_c, u_c, 1.2*u_c]
colors = ['red', 'green', 'blue']
labels = [r'$v < u_c$', r'$v = u_c$', r'$v > u_c$']

L = 1.0
n_points = 200
x = np.linspace(0, L, n_points)

fig = plt.figure(figsize=(16, 7))

ax1 = fig.add_subplot(121, projection='3d')
for i, v in enumerate(speeds):
    F_net = E_mag - v*B_mag
    deflection = 0.5 * F_net * (L/v)**2
    y_path = np.linspace(0, deflection, n_points)
    z_path = np.linspace(0, deflection/2, n_points)
    ax1.plot(x, y_path, z_path, color=colors[i], linewidth=2, label=f"{labels[i]} (v={v:.2f})")

ax1.quiver(0, 0, 0, 0, E_mag/2, 0, color='purple', linewidth=2, length=0.2, normalize=True, label=r'$\mathbf{E}$')
ax1.quiver(0, 0, 0, 0, 0, B_mag/2, color='orange', linewidth=2, length=0.2, normalize=True, label=r'$\mathbf{B}$')
ax1.quiver(0, 0, 0, v_mag, 0, 0, color='cyan', linewidth=2, length=0.2, normalize=True, label=r'$\mathbf{v}$')
ax1.set_title("Velocity Selector in 3D")
ax1.set_xlabel("x-axis")
ax1.set_ylabel("y-axis")
ax1.set_zlabel("z-axis")
ax1.legend(fontsize=9)

ax2 = fig.add_subplot(122)
for i, v in enumerate(speeds):
    F_net = E_mag - v*B_mag
    deflection = 0.5 * F_net * (L/v)**2
    y_path = np.linspace(0, deflection, n_points)
    ax2.plot(x, y_path, color=colors[i], linewidth=2, label=f"{labels[i]} (v={v:.2f})")

ax2.axhline(0, color='black', linestyle='--', label=r"$v = u_c$ line")
ax2.set_title("Velocity Selector in 2D")
ax2.set_xlabel("Distance in Selector (m)")
ax2.set_ylabel("Deflection (arb. units)")
ax2.grid(True, linestyle='--', alpha=0.6)
ax2.legend(fontsize=9)

plt.show()

print(f"Critical speed u_c = {u_c:.2f} m/s")
```
