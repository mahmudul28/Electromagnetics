```python
import numpy as np
import plotly.graph_objects as go

nx, ny, nz = 30, 30, 30
V = np.zeros((nx, ny, nz))
rho = np.zeros((nx, ny, nz))

rho[nx//2, ny//2, nz//2] = 1e-9
epsilon = 8.85e-12

for _ in range(6000):
    V[1:-1, 1:-1, 1:-1] = (V[1:-1, 1:-1, :-2] + V[1:-1, 1:-1, 2:] +
                            V[1:-1, :-2, 1:-1] + V[1:-1, 2:, 1:-1] +
                            V[:-2, 1:-1, 1:-1] + V[2:, 1:-1, 1:-1] +
                            rho[1:-1, 1:-1, 1:-1] / epsilon) / 6

mid = V[:, :, nz//2]

fig = go.Figure(data=[go.Surface(z=mid, colorscale='Inferno')])
fig.update_layout(title='Poissonâ€™s Equation (3D slice with charge)',
                  scene=dict(
                      xaxis_title='x',
                      yaxis_title='y',
                      zaxis_title='Potential (V)',
                  ),
                  width=700, height=600)
fig.show()
```
