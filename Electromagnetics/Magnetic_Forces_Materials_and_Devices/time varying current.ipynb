```python
import numpy as np
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
fig.patch.set_facecolor("white")

titles = [
    "Stationary charge → Electrostatic field",
    "Steady current → Magnetostatic field",
    "Time-varying current → Electromagnetic wave"
]


x, y = np.meshgrid(np.linspace(-2, 2, 25), np.linspace(-2, 2, 25))
r = np.sqrt(x**2 + y**2) + 1e-6
Ex = x / (r**2)
Ey = y / (r**2)
axes[0].quiver(x, y, Ex, Ey, color="blue", pivot="middle", alpha=0.8)
axes[0].scatter(0, 0, color="red", s=40)
axes[0].set_title(titles[0])
axes[0].set_xticks([]); axes[0].set_yticks([])

Bx = -y / (r**2)
By = x / (r**2)
axes[1].quiver(x, y, Bx, By, color="green", pivot="middle", alpha=0.8)
axes[1].plot([0, 0], [-2, 2], color="black", lw=2)
axes[1].set_title(titles[1])
axes[1].set_xticks([]); axes[1].set_yticks([])


ax = axes[2]
t = np.linspace(0, 4*np.pi, 400)
E = np.sin(t)
B = 0.7*np.sin(t + np.pi/2)
ax.plot(t, E, color="blue", lw=2, label="E field →")
ax.plot(t, B, color="red", lw=2, label="B field ↑")
ax.legend()
ax.set_xlim(0, 4*np.pi)
ax.set_ylim(-1.5, 1.5)
ax.set_title(titles[2])
ax.set_xlabel("Space →")
ax.set_ylabel("Field amplitude")

plt.tight_layout()
plt.show()
```
