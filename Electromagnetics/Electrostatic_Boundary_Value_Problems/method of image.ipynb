```python
import numpy as np
import matplotlib.pyplot as plt

R = 1.0          # sphere radius
q = 10          # real charge magnitude
r0 = -2.5         # distance of real charge from center
eps0 = 8.854e-12


q_img = -q * R / r0
r_img = R**2 / r0

x = np.linspace(-3, 3, 400)
y = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x, y)
Z = np.zeros_like(X)


r1 = np.sqrt((X - r0)**2 + Y**2)
r2 = np.sqrt((X - r_img)**2 + Y**2)


V = q / r1 + q_img / r2


V[X**2 + Y**2 < R**2] = np.nan


plt.figure(figsize=(7,6))
cp = plt.contourf(X, Y, V, levels=60, cmap='plasma')
plt.colorbar(cp, label='Potential (arbitrary units)')


circle = plt.Circle((0,0), R, color='white', fill=False, lw=2)
plt.gca().add_patch(circle)


plt.plot(r0, 0, 'wo', label='Real charge (+q)')
plt.plot(r_img, 0, 'ro', label='Image charge (q\')')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Method of Images: Point Charge Near Grounded Conducting Sphere')
plt.legend()
plt.axis('equal')
plt.show()
```
